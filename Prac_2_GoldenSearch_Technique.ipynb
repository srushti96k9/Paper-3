```python
#Golden Search Method - The golden-section search is a technique for finding an extremum (minimum or maximum) of a function inside a specified interval.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func_fx(x):
    return np.sin(x)

def golden_search(xl, xu, mode, et):
    r = (np.sqrt(5) - 1) / 2  # Golden ratio
    it, e = 0, 1

    while e > et:
        d = r * (xu - xl)
        x1, x2 = xu - d, xl + d
        fx1, fx2 = func_fx(x1), func_fx(x2)
        label = 'right' if x2 < x1 else ''

        plt.plot(np.linspace(0, 6, 100), func_fx(np.linspace(0, 6, 100)), label='Function')
        plt.plot(x1, fx1, 'ro', label='x1')
        plt.plot(x2, fx2, 'bo', label='x2')
        plt.title('Golden Section Search')
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.legend()
        plt.show()

        if mode == 'max':
            xl, xu, xopt = (xl, x1, x2) if fx2 > fx1 and label == 'right' else (x2, xu, x1)
        elif mode == 'min':
            xl, xu, xopt = (x2, xu, x1) if fx2 > fx1 and label == 'right' else (xl, x1, x2)
        else:
            print('Please define min/max mode')
            break

        it += 1
        print(f'Iteration: {it}, Error: {e:.4f}')
        e = (1 - r) * abs((xu - xl) / xopt) * 100

# Execute Golden Section Search
golden_search(0, 6, 'max', 0.05)

```
